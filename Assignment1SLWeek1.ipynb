## **Problem 1**

The code demonstrates how the standard error of the mean (SEM) varies with sample size. It generates multiple samples of different sizes from a normal distribution, computes the mean for each sample, calculates the SEM, and then plots the SEM against the sample sizes. It also compares these empirical results with the theoretical SEM, which decreases as the square root of the sample size (1/√N). This type of analysis is common in statistics to understand how sample size affects the precision of estimates derived from sample data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_sizes = [3, 10, 50, 100, 500, 1000] # Define different sample sizes
mean_sds = []# Initialize an empty list to store standard errors of the mean (SEM)
```

```python
# Loop over each sample size N in the list sample_sizes
for N in sample_sizes:
    # Generate 1000 samples of size N from a normal distribution
    # and calculate the mean for each sample
    m = [np.mean(np.random.normal(size=N)) for _ in range(1000)]

    # Calculate the standard deviation (SD) of the means
    sd_m = np.std(m)

    # Append the calculated standard deviation to the list mean_sds
    mean_sds.append(sd_m)
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mean_sds, 'o-', label='SEM')# Plot the SEM (standard error of the mean) against sample sizes
plt.plot(sample_sizes, 1/np.sqrt(sample_sizes), 'b--', label='Théoretical SEM (1/√N)')# Plot the theoretical SEM (1/sqrt(N)) as a dashed line for comparison
# Add labels and title to the plot
plt.xlabel('Sample Size')
plt.ylabel('SEM')
plt.title('SEM vs Sample Size')
plt.legend() # Display the legend to differentiate between plotted lines
plt.grid(True) # Enable grid for better readability of the plot
plt.show()
```


### Conclusion and Interpretation:

The provided code conducts a simulation to explore how the standard error of the mean (SEM) changes with varying sample sizes drawn from a normal distribution. Here are the key takeaways:

1. **SEM Calculation**:
   - SEM is a measure of the precision with which the sample mean estimates the population mean. It decreases as the sample size increases, indicating that larger samples tend to produce more accurate estimates of the population mean.

2. **Empirical vs Theoretical SEM**:
   - The code calculates the empirical SEM by:
     - Drawing 1000 samples of a specified size (`N`) from a normal distribution.
     - Computing the mean of each sample.
     - Calculating the standard deviation of these sample means.
   - It then compares these empirical results with the theoretical SEM (`1/√N`), which predicts how SEM should decrease as the square root of the sample size (`N`).

3. **Plot Interpretation**:
   - The plot visually demonstrates the relationship between sample size (`x-axis`) and SEM (`y-axis`).
   - It shows that as the sample size increases, the SEM decreases, reflecting increased precision in estimating the population mean.
   - The theoretical curve (`1/√N`) serves as a benchmark against which the empirical SEM values are compared.

4. **Statistical Concepts Used**:
   - **Normal Distribution**: Data is generated from a normal distribution, which is a common assumption in statistical analysis.
   - **Sample Mean**: The mean of each sample is computed to estimate the population mean.
   - **Standard Deviation**: Used to quantify the variability of sample means around the true population mean.
   - **Standard Error of the Mean (SEM)**: Indicates the precision of the sample mean estimate and is inversely proportional to the square root of sample size (`1/√N`).

## **Sampling and Observing Distribution Changes**

This exercise involves generating and plotting histograms of sample means for a given sample size
𝑁 from an exponential distribution. The goal is to illustrate the Central Limit Theorem (CLT).

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate and plot histogram
def plot_histogram(N, num_samples=1000):
    # Generate num_samples samples, each of size N, from the exponential distribution
    samples = [np.mean(np.random.exponential(scale=1.0, size=N)) for _ in range(num_samples)]

    # Plot the histogram
    plt.figure(figsize=(10, 6))
    plt.hist(samples, bins=30, density=True, alpha=0.7, color='b')
    plt.title(f'Histogram of sample means for N={N}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

# Plot histogram for N=1
plot_histogram(N=1)
```

For N=1, the histogram represents the exponential distribution itself.

For N=5, the histogram starts to show the effect of the Central Limit Theorem, but it is not quite normal yet.

```python
# Plot histogram for large N
plot_histogram(N=50)
```

For N=50, the histogram shows a distribution that looks very normal, illustrating the Central Limit Theorem in action.

### Interpretation
Plotting Histograms:

The histograms of these sample means are plotted for different values of
𝑁(1, 5, 50).
These histograms show how the distribution of sample means changes as
𝑁 increases.
Central Limit Theorem (CLT)
The Central Limit Theorem states that, given a sufficiently large sample size
𝑁, the distribution of the sample mean will be approximately normally distributed, regardless of the shape of the original distribution, provided that the original distribution has a finite mean and variance.

## **Problem 2**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 1  # the number of i.i.d. variables X we are going to sum
n_exp = 1000  # how many times we are going to repeat the "experiment"

# Initialize an empty list to store sum values in each experiment
s_values = []  # we will use this vector to store the value of the sum in each experiment

# Repeat the experiment n_exp times
for _ in range(n_exp):
    # Step 1: Draw the values x1, ..., xN of the random variables we are going to sum up
    # Here we use the normal distribution as an example
    x = np.random.exponential(scale=1.0, size=N)  # Generate exponential random samples

    # Step 2: Calculate the sum of x1, ..., xN
    s = sum(x)  # Compute the sum

    # Step 3: Save the sum into the vector s_values
    s_values.append(s)

# Draw histogram of n_exp
plt.hist(s_values, bins=30, density=True, alpha=0.75)
plt.xlabel('Sum of i.i.d. Variables')
plt.ylabel('Density')
plt.title('Distribution of Sum of i.i.d. Variables(N=1)')
plt.grid(True)
plt.show()
```

As per the CLT, regardless of the original distribution of the random variables, the distribution of the sum of a large number of such variables tends towards a normal distribution.
