```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres de la grille pour l'illustration
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2) / 2)

# Coordonnes Polaires
theta = np.linspace(0, 2*np.pi, 400)
r = np.linspace(0, 3, 400)
R, Theta = np.meshgrid(r, theta)
X_polar = R * np.cos(Theta)
Y_polar = R * np.sin(Theta)
Z_polar = np.exp(-R**2 / 2)

# Figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Coordonnes Cartsiennes
ax1.contourf(X, Y, Z, levels=50, cmap='viridis')
ax1.set_title('Coordonnes Cartsiennes')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_aspect('equal')

# Coordonnes Polaires
ax2.contourf(X_polar, Y_polar, Z_polar, levels=50, cmap='viridis')
ax2.set_title('Coordonnes Polaires')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```
